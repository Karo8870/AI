```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
submissions = pd.read_csv('./titanic/gender_submission.csv')
test_X = pd.read_csv('./titanic/test.csv')
train = pd.read_csv('./titanic/train.csv')

test_y = submissions.Survived
```

```python
y = train.Survived
# Might need to process cabin
X = train.drop(labels=['Survived', 'Name', 'Ticket', 'Cabin', 'PassengerId', 'Embarked'], axis=1)

X.Sex = X.Sex.map({ 'male': 1, 'female': 0 })
# X.Embarked = X.Embarked.map({'S': 0, 'C': 1, 'Q': 2})
```

```python
model = RandomForestClassifier()

model.fit(X, y)
```

```python
test_X = test_X.drop(['Name', 'Ticket', 'Cabin', 'PassengerId', 'Embarked'], axis=1)

test_X.Sex = test_X.Sex.map({ 'male': 1, 'female': 0 })
# test_X.Embarked = test_X.Embarked.map({'S': 0, 'C': 1, 'Q': 2})

predictions = model.predict(test_X)
```

```python
accuracy_score(predictions, test_y)
```
