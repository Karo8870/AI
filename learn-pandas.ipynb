```python
# Select a column (returns a series)
notes = dataframe.Notes
# or dataframe['Notes']

# Select multiple columns (returns a DataFrame)
some_cols = dataframe[['Notes', 'AppID', 'About the game']]

# Select an item of a series (returns a value)
value = notes[0]

# Index based selection

# Select a row; Works like a python list, so can do partial selecting
row_1 = dataframe.iloc[0]

# Get a column with iloc
col_1 = dataframe.iloc[:, 0]

# Get first 3 rows of a column using iloc
values = dataframe.iloc[:3, 0]

# Select rows based on a list of indexes
other_values = dataframe.iloc[[0, 1, 2], 1]

# label based selection

# Get the country column of the first row
label_values = dataframe.loc[0, 'Notes']

# Get the get more columns of the first row
more_label_values = dataframe.loc[0, ['Notes', 'AppID', 'About the game']]

# Remember that iloc uses stdlib indexing and loc doesn't, so:
some_values = dataframe.loc[0:2, 'AppID']  # Selects rows 0, 1 and 2
some_other_values = dataframe.iloc[0: 2, 0]  # Selects rows 0 and 1

print(notes, value, row_1, col_1, values, other_values, label_values, more_label_values, some_values, some_other_values,
      sep='\n')
```

```python
# Changing the index

dataframe.set_index('AppID')
```

```python
# Conditional selection

# We can use a list of boolean values to select our data. This allows us to use boolean operations to generate lists of booleans to select data.

# Select all rows where Linux is False
dataframe.loc[~dataframe.Linux]

# Select all rows where linux is false and mac is true
dataframe.loc[(dataframe.Linux == False) & (dataframe.Mac == True)]

# Select all rows that have either 2 or 3 DLCs
dataframe.loc[dataframe['DLC count'].isin([2, 3])]

# Select all rows where price is not null (there are also isna and isnull and notna)
dataframe.loc[dataframe['Price'].notnull()]
```

```python
# Assigning data

# Assigns the price 0 to all rows
dataframe.Reviews = 0

# Assigns reverse indexes
dataframe.AppID = range(len(dataframe), 0, -1)
```

```python
# Summary functions

# Describing a dataframe
dataframe.describe()

# Describing a Series (text and numeric)
dataframe.Notes.describe()
dataframe.Price.describe()

# Getting the first items
dataframe.head(5)

# Getting the average / middle value
dataframe['Price'].mean()
dataframe['Price'].median()

# Getting unique values
dataframe['Price'].unique()

# Getting the count of all unique values
dataframe['Price'].value_counts()

# Maps are used to map values in a series to other values
mean = dataframe['Price'].mean()
dataframe['Price'].map(lambda price: price - mean)


# Apply is used to transform a dataframe (one row at a time)
def re_mean(row):
    row.Price = row.Price - mean
    
    return row


dataframe.apply(re_mean, axis='columns')  # If we specified index instead of columns, we would modify columns

# # Get the index of the maximum (might be useful in combination with other operations)
# dataframe.Price.idxmax()
# 
# # Instead of traditional counting, you can count boolean values by summing them up, ex: how many games have prices below 50?
# dataframe['Price'].map(lambda price: price < 50).sum()
```

```python
# Grouping

# Grouping allows us to group dataframes by one or more columns into a new data type called DataFrameGroupBy (It also has the Series equivalent)

# This is the equivalent of value_counts() method
dataframe.groupby(['Release date', 'Price']).count()

# You can get the minimum price of a release date by using:
dataframe.groupby('Release date').Price.min()

# We can also manipulate groups using apply as we would edit a DataFrame, except that here we are mapping over sub-DataFrames instead of rows
dataframe.groupby('Price').apply(lambda df: df['Release date'].iloc[0])

# We can apply aggregation functions on groups using agg. You can pass either predefined strings or aggregation functions
dataframe.groupby('Release date')['Price'].agg(['max', 'min', len, 'sum', 'count'])

# groupby may sometimes result in a multi-index (a dataframe with multiple indexes)
multiple_index = dataframe.groupby(['Release date', 'Price']).count()

# You may go back to normal indexing by calling reset_index
multiple_index.reset_index()
```

```python
# Sorting

# We can use the sort_values method to sort both dataframes and multi index dataframes. You can sort by index or by value
dataframe.sort_values(by='Release date')
multiple_index.sort_values(by='AppID', ascending=False)
dataframe.sort_index()
dataframe.sort_values(by=['AppID', 'Price'], ascending=[True, False])

```

```python
# Data types and missing values

# All columns have a datatype ex. float64, object etc.
dataframe.dtypes
dataframe.Price.dtype

# To convert a column to another type we can use the astype method
dataframe.Price.astype('float32')

# Index columns have dtypes too
dataframe.index.dtype

# NaN is always dtype('float64')

# We can fill unknown data using fillna
dataframe['Price'][2] = None
dataframe['Price'].fillna('Unknown')

# We can also replace data
dataframe['Price'].replace(19.99, 20)
```

```python
# Renaming

# We can rename columns ex. Name to not_name
dataframe.rename(columns={'Name': 'not_name'})

# We can also rename row indexes
dataframe.rename(index={0: 'first_col', 1: 'second_col'})

# Both row index and column index can have a name attribute. We use rename_axis for those
dataframe.rename_axis('steam games', axis='rows').rename_axis('props', axis='columns')
```

```python
# Combining

df1 = pd.DataFrame({
    'col_1': ['a', 'b', 'c'],
    'col_2': ['d', 'e', 'f'],
    'shared_col': [1, 2, 3]
})

df2 = pd.DataFrame({
    'col_3': ['g', 'h', 'i'],
    'col_4': ['j', 'k', 'l'],
    'shared_col': [1, 2, 3]
})

# The first method is concat, allowing us to insert a dataframe after another
pd.concat([df1, df2])

# The second method is join, allowing us to merge rows together by index
df1.join(df2, lsuffix='1', rsuffix='2')  # We must specify lsuffix and rsuffix when there are overlapping names
```

```python
# Deleting data

# Data is deleted using the drop method

# Drop a row by index
dataframe.drop(0)

# Delete data using conditional filtering
dataframe[dataframe['Price'] < 1]

# Deleting rows
dataframe.drop(columns=['Price', 'AppID', 'Release date'])

# Deleting duplicates
dataframe.drop_duplicates(subset=['Price', 'Release date'])

# Deleting missing data
dataframe.dropna(how='any', subset=['Price'])
```
